```python
df_2017=pd.read_csv("use_data/일평균_추정교통량(2017).csv", encoding='utf-8-sig')
df_2018=pd.read_csv("use_data/일평균_추정교통량(2018).csv", encoding='utf-8-sig')
```

```python
# 의미있는 행 추출
df_truck=df_2017[['시도','시군구','트럭-평일']]
```

```python
df_truck=df_truck.rename(columns={'트럭-평일':'2017_트럭일평균수'})
```

```python
df_truck=pd.concat([df_truck,df_2018['트럭-평일']], axis=1)
```

```python
df_truck=df_truck.rename(columns={'트럭-평일':'2018_트럭일평균수'})
```

```python
df_truck.head()
```

```python
df_truck.reset_index(inplace=True)
```

```python
df_truck
```

```python
df_truck['평균']=np.round((df_truck['2017_트럭일평균수']+df_truck['2018_트럭일평균수'])/2,0).astype(int)
```

```python
df_truck.head()
```

```python
df_truck.query("시도=='경기도'")
```

```python
df_truck['ZONE ID']=df_truck['ZONE ID'].astype('str')
```

```python
zone=list(df_truck['ZONE ID'].map(lambda x: x[:4]+'000'))
```

```python
zone=list(set(zone))
```

```python
zone.sort()
```

```python
df_truck_r=pd.DataFrame(columns=df_truck.columns)
df_truck_r['ZONE ID']=zone
df_truck_r=df_truck_r.set_index('ZONE ID')
temp={}
for i in range(len(df_truck)):
    zone_df=df_truck.loc[i,'ZONE ID'][:4]+'000'   
    if df_truck.loc[i,'ZONE ID'][4]!='0':
        if df_truck.loc[i,'ZONE ID'][:4]+'000' not in temp.keys():
            temp[df_truck.loc[i,'ZONE ID'][:4]+'000']=[]
        temp[zone_df].append([df_truck.loc[i,'2017_트럭일평균수'],df_truck.loc[i,'2018_트럭일평균수'],df_truck.loc[i,'평균']])
    else:
        df_truck_r.loc[zone_df,'2017_트럭일평균수']=df_truck.loc[i,'2017_트럭일평균수']
        df_truck_r.loc[zone_df,'2018_트럭일평균수']=df_truck.loc[i,'2018_트럭일평균수']
        df_truck_r.loc[zone_df,'평균']=df_truck.loc[i,'평균']
    df_truck_r.loc[zone_df,'시도']=df_truck.loc[i,'시도']
    df_truck_r.loc[zone_df,'시군구']=df_truck.loc[i,'시군구'].split()[0]
```

```python
df_truck_r.query('시도=="경기도"')
```

```python
temp
```

```python
for index in temp.keys():
    truck_2017=0
    truck_2018=0
    mean=0
    cnt=len(temp[index])
    for lst in temp[index]:
        truck_2017+=lst[0]
        truck_2018+=lst[1]
        mean+=lst[2]
    truck_2017=int(np.round(truck_2017/cnt,0))
    truck_2018=int(np.round(truck_2018/cnt,0))
    mean=int(np.round(mean/cnt,0))
    df_truck_r.loc[index,'2017_트럭일평균수']=truck_2017
    df_truck_r.loc[index,'2018_트럭일평균수']=truck_2018
    df_truck_r.loc[index,'평균']=mean
```

```python
# 2017~2018 평균 발생건수
data=pd.DataFrame(columns=['시도','시군구','평균발생건수'])
data['시도']=sido['시도']
data['시군구']=sido['시군구']
data['평균발생건수']=0
data=data.set_index(['시도','시군구'])
dic={}
for i in range(len(occur_hpai)):
    if occur_hpai.loc[i,'year']==2017 or occur_hpai.loc[i,'year']==2018:
        address=occur_hpai.loc[i,'시도']+' '+occur_hpai.loc[i,'시군구']
        if address not in dic.keys():
            dic[address]=0
        else:
            dic[address]+=1
```

```python
for key in dic.keys():
    temp=key.split()
    data.loc[(temp[0],temp[1]), '평균발생건수']=np.round(dic[key]/2,1)
```

```python
data.reset_index(inplace=True)
```

```python
data['truck_mean']=0
for i in range(len(data)):
    for j in range(len(df_truck_r)):
        if df_truck_r.loc[j,'시도']==data.loc[i,'시도'] and df_truck_r.loc[j,'시군구']==data.loc[i,'시군구']:
            data.loc[i,'truck_mean']=df_truck_r.loc[j,'평균']
            break
```

```python
data['면적']=np.nan
for i in range(len(data)):
    k=0
    for j in range(len(m)):
        if m.loc[j,'시도']==data.loc[i,'시도'] and data.loc[j,'시군구']=='세종시':
            data.loc[i,'면적']=m.loc[j,'2018']
            k+=1
            break
        if m.loc[j,'시도']==data.loc[i,'시도'] and m.loc[j,'시군구']==data.loc[i,'시군구']:
            data.loc[i,'면적']=m.loc[j,'2018']
            k+=1
            break
        if m.loc[j,'시도']==data.loc[i,'시도'] and data.loc[i,'시군구']=='남구':
            data.loc[i,'면적']=m.loc[j,'2018']
            k+=1
            break
    if k==0:
        print(data.loc[i,'시도'],data.loc[i,'시군구'])
```

```python
import matplotlib.pyplot as plt
plt.scatter(use_data['트럭밀집도'],use_data['평균발생건수'])
```

```python
from scipy import stats
```

```python
# 그래프
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(3,2,1)
ax2 = fig.add_subplot(3,2,2)
ax3 = fig.add_subplot(3,2,3)
ax4 = fig.add_subplot(3,2,4)
ax5 = fig.add_subplot(3,2,5)
ax6 = fig.add_subplot(3,2,6)

data=use_not_data['트럭밀집도'] # Q-Q plot
stats.probplot(data, dist=stats.hypsecant, plot=ax1)
ax1.legend(['hypsecant'])
stats.probplot(data, dist=stats.uniform(0,2), plot=ax2)
ax2.legend(['uniform(0,2)'])
stats.probplot(data, dist=stats.norm, plot=ax3)
ax3.legend(['normal'])
stats.probplot(data, dist=stats.t(15), plot=ax4)
ax4.legend(['t(15)'])
stats.probplot(data, dist=stats.f(15,15), plot=ax5)
ax5.legend(['f(15,15)'])
stats.probplot(data, dist=stats.lognorm(1), plot=ax6)
ax6.legend(['lognorm(1)'])
plt.show()
```

```python
stats.probplot(use_data['트럭밀집도'], dist=stats.uniform(0,2), plot=plt)
plt.show()
```

```python
stats.probplot(use_not_data['트럭밀집도'], dist=stats.norm, plot=plt)
plt.show()
```

```python
#정규성검정
test_stat, p_value = stats.shapiro(use_data['트럭밀집도'])
print("test_stat: %f, pvalue=%f"%(test_stat, p_value))
test_stat2, p_value2 = stats.shapiro(use_not_data['트럭밀집도'])
print("test_stat: %f, pvalue=%f"%(test_stat2, p_value2))
```

```python
# 카이제곱 검정
crosstab=pd.crosstab(use_data['트럭밀집도'],use_not_data['트럭밀집도'])
chi2 = stats.chi2_contingency(crosstab)
print("test_stat: %f, pvalue=%f"%(chi2[0], chi2[1]))
```

```python
# u검정
result_temp=stats.mannwhitneyu(use_data['트럭밀집도'],use_not_data['트럭밀집도'])
print("검정 통계량: %.3f, pvalue=%.3f"%(result_temp))
```
