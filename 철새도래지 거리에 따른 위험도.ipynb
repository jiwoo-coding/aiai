```python
birds=pd.read_csv('철새도래지_테스트용.csv')
birds
```

```python
x=birds['위험도 점수']
y=birds['거리']
xy=pd.concat([x,y], axis=1)
xy
```

```python
#정규화 minmax 하기
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(xy)
Normalized_Values = scaler.transform(xy)
```

```python
Normalized_Values
```

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=11, random_state =0).fit(Normalized_Values)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
xy
```

```python
sb.lmplot('위험도 점수', '거리', data=xy, fit_reg=False, scatter_kws={"s": 100})
plt.title('k-means Example')
plt.xlabel('위험도 점수')
plt.ylabel('거리')
```

```python
points=Normalized_Values
kmeans=KMeans(n_clusters=11).fit(points)

kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
xy['risk']=kmeans.labels_
xy.loc[200:250]
```

```python
xy.to_csv("철새도래지_위험도_테스트.csv", encoding='utf-8-sig')
```
